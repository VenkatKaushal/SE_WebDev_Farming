```python
from sklearn.model_selection import train_test_split
from xgboost import XGBRFRegressor
from sklearn.pipeline import Pipeline
from sklearn.metrics import mean_absolute_error
import matplotlib.pyplot as plt
```

```python
X_train, X_valid, y_train, y_valid= train_test_split(X, y, random_state=1)
```

```python
best_model = XGBRFRegressor(random_state=1)
```

```python
def argmin(lst):
  return lst.index(min(lst))
def get_learn_rate(n1,n2,n3):
    lista = [i/100 for i in range(n1,n2,n3)]

    results = dict()

    for n in lista:
        best_model.set_params(n_estimators=800,
                              learning_rate=n,
                              n_jobs=-1,
                              random_state=10)

        modelo_x = best_model.fit(X_train, y_train)
        prediction = modelo_x.predict(X_valid)
        results[n] = mean_absolute_error(y_valid, prediction)

    for x,scores in results.items():
        print('parameter:', x)
        print(f'MAE:{scores}\n')

    plt.plot(list(results.keys()), list(results.values()))
    plt.show()
    scores_list = [x for x in results.values()]
    min_index = int(argmin(scores_list))
    learn = [x for x in results.keys()]
    return learn[min_index]
```

```python
learning_rate = get_learn_rate(98,102,1)
```

```python
def get_n_estimators(n1,n2,n3):
    lista = [i for i in range(n1,n2,n3)]

    results = dict()

    for n in lista:
        best_model.set_params(n_estimators=n,
                              learning_rate=learning_rate,
                              n_jobs=-1,
                              random_state=10)

        modelo_x = best_model.fit(X_train, y_train)
        prediction = modelo_x.predict(X_valid)
        results[n] = mean_absolute_error(y_valid, prediction)

    for x,scores in results.items():
        print('parameter:', x)
        print(f'MAE:{scores}\n')

    plt.plot(list(results.keys()), list(results.values()))
    plt.show()
```

```python
get_n_estimators(750,850,10)
```

```python
def get_colsample_bynode(n1,n2,n3):
    lista = [i/100 for i in range(n1,n2,n3)]

    results = dict()

    for n in lista:
        best_model.set_params(n_estimators=800,
                              learning_rate=1,
                              colsample_bynode=n,
                              n_jobs=-1,
                              random_state=10)

        modelo_x = best_model.fit(X_train, y_train)
        prediction = modelo_x.predict(X_valid)
        results[n] = mean_absolute_error(y_valid, prediction)

    for x,scores in results.items():
        print('parameter:', x)
        print(f'MAE:{scores}\n')

    plt.plot(list(results.keys()), list(results.values()))
    plt.show()
```

```python
get_colsample_bynode(30,40,3)
```

```python
def get_max_depth(n1,n2,n3):
    lista = [i for i in range(n1,n2,n3)]

    results = dict()

    for n in lista:
        best_model.set_params(n_estimators=800,
                              learning_rate=1,
                              colsample_bynode=0.3,
                              max_depth=n,
                              n_jobs=-1,
                              random_state=10)

        modelo_x = best_model.fit(X_train, y_train)
        prediction = modelo_x.predict(X_valid)
        results[n] = mean_absolute_error(y_valid, prediction)

    for x,scores in results.items():
        print('parameter:', x)
        print(f'MAE:{scores}\n')

    plt.plot(list(results.keys()), list(results.values()))
    plt.show()
```

```python
get_max_depth(12,25,1)
```

```python
def get_paramet(n1,n2,n3):
    lista = [i for i in range(n1,n2,n3)]

    results = dict()

    for n in lista:
        best_model.set_params(n_estimators=800,
                              learning_rate=1,
                              colsample_bynode=0.3,
                              max_depth=n,
                              n_jobs=-1,
                              random_state=10)

        modelo_x = best_model.fit(X_train, y_train)
        prediction = modelo_x.predict(X_valid)
        results[n] = mean_absolute_error(y_valid, prediction)

    for x,scores in results.items():
        print('parameter:', x)
        print(f'MAE:{scores}\n')

    plt.plot(list(results.keys()), list(results.values()))
    plt.show()
```

```python
get_paramet(12,15,1)
```

```python
rf_model = best_model.set_params(n_estimators=800,
                              learning_rate=1,
                              colsample_bynode=0.3,
                              max_depth=13,
                              n_jobs=-1,
                              random_state=10)
rf_model.fit(X_train, y_train)
predict = rf_model.predict(X_valid)
```
